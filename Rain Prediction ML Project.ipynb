```python
df=df.drop(['Evaporation','Sunshine','Date'],axis=1)
df.head()
```

```python
df=df.dropna(axis=0)
df.shape
```

```python
#Fill null values
df.fillna(df.mode().iloc[0])
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
# Drop rows with missing target values (Update 'RainTomorrow' with your target variable)
df = df.dropna(subset=['RainTomorrow'])

# Separate features and target (Update 'RainTomorrow' with your target variable)
X = df.drop(columns=['RainTomorrow'])  # Update 'Date' if it's irrelevant to your case
y = df['RainTomorrow']
# Identify numerical and categorical columns
num_cols = X.select_dtypes(include=['float64']).columns
cat_cols = X.select_dtypes(include=['object']).columns

# Pipeline for numerical features: impute missing values and scale
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', MinMaxScaler())
])

# ColumnTransformer to handle both numerical and categorical data
preprocessor = ColumnTransformer([
    ('num', num_pipeline, num_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
])

# Apply the transformations
X_preprocessed = preprocessor.fit_transform(X)
```

```python
X_preprocessed
```

```python
from sklearn.feature_selection import SelectKBest, chi2
selector = SelectKBest(score_func=chi2, k=10)
X_selected = selector.fit_transform(X_preprocessed, y)
```

```python
def get_feature_names(cat_cols, preprocessor, num_cols):
    cat_features = preprocessor.named_transformers_['cat']
    cat_feature_names = cat_features.get_feature_names(cat_cols)
    return np.concatenate([num_cols, cat_feature_names])
feature_names = get_feature_names(cat_cols, preprocessor, num_cols)
selected_mask = selector.get_support()

# Get the names of the selected features
selected_features = feature_names[selected_mask]

print("Selected Features:")
print(selected_features)
```

```python
from scipy import stats

z_scores=stats.zscore(df[['Temp3pm','Humidity3pm','WindSpeed3pm']])

threshold=3
outliers=(abs(z_scores)>threshold).any(axis=1)

# Remove outliers
data_cleaned_zscore=df[~outliers]

print(f'Removed {outliers.sum()} outliers using Z-score method')
```

```python
Q1=df[['Temp3pm','Humidity3pm','WindSpeed3pm']].quantile(0.25)
Q3=df[['Temp3pm','Humidity3pm','WindSpeed3pm']].quantile(0.75)
IQR=Q3-Q1

outlier_mask=((df[['Temp3pm','Humidity3pm','WindSpeed3pm']]<(Q1 - 1.5 * IQR)) |
              (df[['Temp3pm','Humidity3pm','WindSpeed3pm']]>(Q3 + 1.5 * IQR)))

data_cleaned_iqr=df[outlier_mask.any(axis=1)]
print(data_cleaned_iqr)
```
